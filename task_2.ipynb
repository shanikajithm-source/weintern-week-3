```python
import pandas as pd

df = pd.read_csv('/content/cleaned_superstore.csv')

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['sales'], bins=30)
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

```python
category_sales = df.groupby('category')['sales'].sum()

category_sales.plot(kind='bar')

plt.title("Total Sales by Category")
plt.ylabel("Sales")
plt.show()
```

```python
numeric_cols = ['sales','quantity','discount','profit']

sns.heatmap(df[numeric_cols].corr(),
            annot=True)

plt.title("Correlation Matrix")
plt.show()
```

```python
sns.scatterplot(x='discount',
                y='sales',
                data=df)

plt.title("Discount vs Sales")
plt.show()
```

```python
sns.boxplot(x='category',
            y='profit',
            data=df)

plt.title("Profit Distribution by Category")
plt.show()
```

Key Insights

1. Technology products contribute a significant portion of total sales compared to other categories.

2. Sales distribution is highly skewed, indicating that a small number of transactions generate very large sales values.

3. Discounts do not always result in higher sales, suggesting that discount strategies should be carefully evaluated.

4. Profitability varies across categories, with some products showing large profit fluctuations.

5. Most customer orders contain relatively small quantities, indicating frequent low-volume purchases.

Conclusion

The exploratory data analysis revealed important patterns in sales, profit, quantity, and discount behavior. Technology products generate strong sales performance, while profitability varies significantly across categories. The analysis provides useful business insights that can help improve sales strategies, inventory management, and promotional planning.
